```python
import folium

# Criar um objeto de mapa
m = folium.Map(location=[-14.235004, -51.92528], zoom_start=4)

# Criar camadas principais (pais)
camada_pontos_interesse = folium.FeatureGroup(name='Pontos de Interesse',show=False)
camada_hotel = folium.FeatureGroup(name='Hotéis',show=False)
camada_restaurante = folium.FeatureGroup(name='Restaurantes',show=False)

# Adicionar marcadores às camadas principais
# Suponha que você tenha dados de lat/long para pontos de interesse, hotéis e restaurantes
# Adicione marcadores para pontos de interesse
folium.Marker([-15.7801, -47.9292], popup='Ponto de Interesse A').add_to(camada_pontos_interesse)
folium.Marker([-22.9068, -43.1729], popup='Ponto de Interesse B').add_to(camada_pontos_interesse)

# Adicione marcadores para hotéis
folium.Marker([-15.7801, -47.9292], popup='Hotel A').add_to(camada_hotel)
folium.Marker([-22.9068, -43.1729], popup='Hotel B').add_to(camada_hotel)

# Adicione marcadores para restaurantes
folium.Marker([-23.5505, -46.6333], popup='Restaurante X').add_to(camada_restaurante)
folium.Marker([-30.0277, -51.2287], popup='Restaurante Y').add_to(camada_restaurante)

# Adicione as camadas principais (pais) ao mapa
camada_pontos_interesse.add_to(m)
camada_hotel.add_to(m)
camada_restaurante.add_to(m)

# Adicione camadas secundárias (filhas)
camada_hotel_chique = folium.FeatureGroup(name='Hotéis Chiques', show=False)
camada_hotel_economico = folium.FeatureGroup(name='Hotéis Econômicos', show=False)

# Adicione marcadores para hotéis chiques
folium.Marker([-15.7801, -47.9292], popup='Hotel 5 Estrelas').add_to(camada_hotel_chique)
folium.Marker([-22.9068, -43.1729], popup='Hotel de Luxo').add_to(camada_hotel_chique)

# Adicione marcadores para hotéis econômicos
folium.Marker([-15.7801, -47.9292], popup='Hotel Econômico A').add_to(camada_hotel_economico)
folium.Marker([-22.9068, -43.1729], popup='Hotel Econômico B').add_to(camada_hotel_economico)

# Adicione as camadas secundárias (filhas) às camadas principais (pais)
camada_hotel_chique.add_to(camada_hotel)
camada_hotel_economico.add_to(camada_hotel)

# Adicione controles de camadas aninhadas (nested layer control)
folium.LayerControl(collapsed=False).add_to(m)

# Salve o mapa em um arquivo HTML
m.save('mapa_com_sublayers.html')
m
```
